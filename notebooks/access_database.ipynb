```python
import numpy as np
import pandas as pd
import psycopg
```

```python
def get_dataframe(table: str) -> pd.DataFrame:
    with psycopg.connect("dbname=zapisy") as conn:
        with conn.cursor() as cur:
            cur.execute(f"SELECT * FROM {table};")
            columns = [desc[0] for desc in cur.description]
            records = cur.fetchall()
    return pd.DataFrame(records, columns=columns)
```

```python
course_instance_pdf = get_dataframe('courses_courseinstance')
```

```python
proposal_pdf = get_dataframe('proposal_proposal')
```

```python
course_info_pdf = get_dataframe('courses_courseinformation')
```

```python
proposal_ids = proposal_pdf['courseinformation_ptr_id'].values
proposals = course_info_pdf[course_info_pdf['id'].isin(proposal_ids)]
```
